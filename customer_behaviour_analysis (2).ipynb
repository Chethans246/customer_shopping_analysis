## Hypothesis Testing

We test three business relevant hypotheses using statistical tests to validate or challenge assumptions in the data.

---

### Hypothesis 1 Do discount users spend more than non-discount users?
**H₀:** Mean purchase amount is the same regardless of discount usage  
**H₁:** Customers who used a discount spend more on average  
**Test:** Independent samples t-test (two-tailed, α = 0.05)

```python
from scipy import stats

discount_yes = df[df['discount_applied'] == 'Yes']['purchase_amount']
discount_no  = df[df['discount_applied'] == 'No']['purchase_amount']

t_stat, p_val = stats.ttest_ind(discount_yes, discount_no, equal_var=False)

print(f"discount group Mean spend: ${discount_yes.mean():.2f}, n={len(discount_yes)}")
print(f"no discount group Mean spend: ${discount_no.mean():.2f}, n={len(discount_no)}")
print(f"\nwelch's t-statistic : {t_stat:.4f}")
print(f"p-value             : {p_val:.4f}")
print()
if p_val < 0.05:
    print("Result: REJECT H₀ -There is a statistically significant difference in spend between discount and non-discount customers.")
else:
    print("Result: FAIL TO REJECT H₀ -No statistically significant difference in spend between the two groups.")
```

Discount customers and non-discount customers spend almost the same amount on average ( 59.28 vs $60.13). Discounts are not driving higher transaction value in this dataset

---

### Hypothesis 2 Do younger customers (young_adults) buy more frequently than seniors?
**H₀:** Median purchase frequency (days between purchases) is the same for young adults and seniors  
**H₁:** Young adults purchase more frequently (fewer days between purchases)  
**Test:** Mann-Whitney U (non-parametric, since frequency_days is ordinal-mapped)

```python
young = df[df['age_group'] == 'young_adults']['Purchase_frequency_days']
senior = df[df['age_group'] == 'senior']['Purchase_frequency_days']

u_stat, p_val2 = stats.mannwhitneyu(young, senior, alternative='less')

print(f"young adults Median frequency: {young.median():.0f} days, n={len(young)}")
print(f"seniors Median frequency: {senior.median():.0f} days, n={len(senior)}")
print(f"\nmann-whitney U statistic : {u_stat:.0f}")
print(f"p-value                  : {p_val2:.4f}")
print()
if p_val2 < 0.05:
    print("Result: REJECT H₀ -Young adults purchase significantly more frequently than seniors.")
else:
    print("Result: FAIL TO REJECT H₀ - No significant difference in purchase frequency between age groups.")
```

Gender does not significantly predict product category preference in this dataset. Marketing segmentation by gender may not be effective for this retailer.

---

### Hypothesis 3 Is there an association between gender and product category preference?
**H₀:** Gender and product category are independent  
**H₁:** There is a statistically significant association between gender and category  
**Test:** Chi-square test of independence (α = 0.05)

```python
contingency = pd.crosstab(df['gender'], df['category'])
chi2, p_val3, dof, expected = stats.chi2_contingency(contingency)

print("Observed counts (Gender × Category):")
print(contingency)
print(f"\nChi-square statistic : {chi2:.4f}")
print(f"Degrees of freedom   : {dof}")
print(f"p-value              : {p_val3:.4f}")
print()
if p_val3 < 0.05:
    print("Result: REJECT H₀ — Gender and product category are NOT independent; purchasing patterns differ by gender.")
else:
    print("Result: FAIL TO REJECT H₀ — No significant association between gender and product category.")
```

All three hypotheses were not supported by the data. This suggests demographic variables alone are insufficient for segmentation a clustering approach on behavioral variables (spend, frequency, category) would likely yield more actionable segments."
